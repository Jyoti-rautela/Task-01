## Task 02: Categories (Sets)

```python
# Categories list
categories = ['Stationery','Clothing','Clothing','Household','Stationery','Clothing','Food','Clothing']
```

```python
# making categories list-> set
categories_set = set(categories)
print(categories_set)
```

```python
# New category added
categories_set.add('Electronics')
print(categories_set)
```

```python
# Duplicate are ignored
print("All categories: ",categories)
print("Duplicates are ignored: ",categories_set)
```

```python
# Checking category in set in boolean format
is_found =  'Food' in categories_set
print(is_found)
```

```python
# Total number of Unique categories using a set
Total = len(categories_set)
print("Total number of Unique categories are:",Total)
```

## Task 03 : Product Pricing (Dictionaries)

```python
price_dict = {'Phone': 18750.78 , 'Pen': 10, 'Jeans': 850, 'Book': 450, 'Bottle': 375.6 , 'Colors': 200.05}
```

```python
# Adding new product to price_dict
price_dict['Copy'] = 50
print(price_dict)
```

```python
# Update existing prioce
price_dict['Colors'] = 150
print(price_dict)
```

```python
# Removing product by name and handling when item isn't there in dictionary

price_dict = {'Phone': 18750.78 , 'Pen': 10, 'Jeans': 850, 'Book': 450, 'Bottle': 375.6 , 'Colors': 200.05}
item = input('enter price_dict item name you want to remove: ')
if item in price_dict:
    del price_dict[item]   # Pen key-value is removed
    print(price_dict)
else:
    print('Item not found in dictionary')
```

```python
# Average price of all product 
Total = sum(price_dict.values())
length = len(price_dict)
Average = Total/length
print("Average price of all product are:",Average)
```

```python
# Product with both max and min prices
maximum = max(price_dict, key=price_dict.get)
print(maximum)
minimum = min(price_dict, key=price_dict.get) 
print(minimum)
```

## Task 04: Combined Operations

```python
# Creating Catalog
products = ['Book','Headphone','Pen','Phone','Maggie','Jeans', 'Ramen', 'Icecream','Hoody']
price_dict = {'Book': 469 , 
              'Headphone': 2045, 
              'Pen': 10, 
              'Phone':25000.78,  
              'Maggie': 50.32, 
              'Jeans': 850, 
              'Ramen': 30, 
              'Icecream': 100, 
              'Hoody': 999.99
              }

categories = ['Stationary', 'Electronics', 'Stationary', 'Electronics','Food','Clothes','Food','Food','Clothes']

catalog = []
for i in range(len(products)):
    product_name = products[i]
    price = price_dict[product_name]
    category = categories[i]
    catalog.append((product_name,price,category))
print(catalog)
```

```python
# category_to_products thet maps eaach category to a list of product name in that category:

category_to_products = {}
for product_name,price,category in catalog:
    if category not in category_to_products:
        category_to_products[category] = []
    category_to_products[category].append(product_name)
    
print(category_to_products)
```

```python
# printing all product of that category whose products is max
max_category = max(category_to_products,key=lambda key:len(category_to_products[key]))
print('Category with maximum products:', max_category)
print('Products:',category_to_products[max_category])
```
